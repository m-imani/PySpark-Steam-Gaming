```python
################Join all `Games_` tables into one dataframe##################

from pyspark.sql import SQLContext
# first of all, I create an empty StructType
schema = StructType([])

#Create an empty dataframe based on the schema
joinedDF=sqlContext.createDataFrame(sc.emptyRDD(), schema)

#inner-join games_1Dataframe with games_2Dataframe based on their common column app_id and then assign the result to empty dataframe 
joinedDF=games_1Dataframe.join(games_2Dataframe, ['appid'])

#inner-join the result(joinedDF) with games_genresDataframe and assign to joinedDF
joinedDF=joinedDF.join(games_genresDataframe, ['appid'])

#inner-join the result(joinedDF) with games_publisherDataframe and assign to joinedDF
joinedDF=joinedDF.join(games_publishersDataframe, ['appid'])

#inner-join the result(joinedDF) with games_developerDataframe and assing the final result to joinedDF
joinedDF=joinedDF.join(games_developersDataframe,['appid'])

# show final result
joinedDF.show(10)


```

```python
####Find day and hour when most new accounts were created (based on Player_Summaries table) e.g. 8pm on 14th August 2005.#####
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *

#create session and read csv file
spark = SparkSession.builder.getOrCreate()
Player_Summariespath="c:\dataset\Player_Summaries.csv"
Player_SummariesDF=spark.read.csv(Player_Summariespath, header=True)

#sort column timecreated from newest to oldest
sorted_Player_SummariesDF = Player_SummariesDF.orderBy('timecreated',ascending=False) 

#get timecreated column from the dataframe
sorted_Player_SummariesDF_time = sorted_Player_SummariesDF.select("timecreated")

#check dataframe schema to make sure the above line works
sorted_Player_SummariesDF_time.printSchema()


#splitting the timestamp column to 2 columns (date, hour)
Player_SummariesDF_timestamp=sorted_Player_SummariesDF_time.select(
    to_date('timecreated').alias('date'),
    hour("timecreated").alias('hour')
)

#check dataframe schema to make sure spliting works
Player_SummariesDF_timestamp.printSchema()

# group dataframe by date to find the most created account date
Player_SummariesDF_grouped=Player_SummariesDF_timestamp.groupby('date').agg({'date': 'count'})

#order by the results descending
Player_SummariesDF_ordered=Player_SummariesDF_grouped.orderBy('count(date)', ascending=False)

# show the 10 first rows
Player_SummariesDF_ordered.show(10)

#find the most created account date using max
Player_SummariesDF_max=Player_SummariesDF_ordered.agg({'count(date)':'max'})

print('Max number of accounts created in one day is:')
Player_SummariesDF_max.show()

#this time, sort the dataframe by date to find the the most created account date
Player_SummariesDF_date=Player_SummariesDF_ordered.select(Player_SummariesDF_ordered['date'])

# pick the first row of the sorted dataframe (the most created account date)
Player_SummariesDF_first=Player_SummariesDF_date.first()

#compare the date column in the original dataframe with the most created account date 
#to filter other dates and assign it to a new dataframe 
Day_Most_Created=Player_SummariesDF_timestamp.where(Player_SummariesDF_timestamp.date == Player_SummariesDF_first.date)

#this time, group the dataframe by hour to find the most created account hour
Day_Most_Grouped=Day_Most_Created.groupby('hour').agg({'hour':'count'})

#sort the results descending
Hour_Most_Created=Day_Most_Grouped.orderBy('count(hour)', ascending=False)
Hour_Most_Created.show(10)

#now, find the most created account hour using max
Most_Created=Day_Most_Grouped.agg({'count(hour)':'max'})

print('Max number of accounts created in one hour in that specified date is:')
Most_Created.show()

#get data frame column value as variable
most_hour = Hour_Most_Created.select('hour').collect()[0].asDict()['hour']
print('The hour when most new accounts were created:', most_hour)

#get data frame column value as variable
most_date = Day_Most_Created.select('date').collect()[0].asDict()['date']
print('The date when most new accounts were created:', most_date)

```
